## Yearly Gas production

```python
yearly_gas_df = gas_df.groupby(gas_df["Month"].dt.year).sum(numeric_only=True)
yearly_gas_df
```


### Check your code

```python
index_year = yearly_gas_df.index[0]
yearly_gas_shape = yearly_gas_df.shape
us_total = yearly_gas_df.iloc[0,0]
```

```python
from nbresult import ChallengeResult

result = ChallengeResult('full_gas',
    index_year=index_year,
    yearly_gas_shape=yearly_gas_shape,
    us_total=us_total
)
result.write()

print(result.check())
```

```python
plot = yearly_gas_df.filter(items=['U.S.']).plot(kind="bar")
plot.set_xlabel("Year");
```

```python
gas_df["Month"] = pd.to_datetime(gas_df["Month"])
monthly_counts = gas_df["Month"].dt.year.value_counts().sort_index()
monthly_counts
```

```python
years = yearly_gas_df.index

filtered_yearly_gas_df = yearly_gas_df[
    np.logical_and(
        years != years.min(),
        years != years.max()
    )
]
filtered_yearly_gas_df
```

```python
filtered_yearly_gas_df.plot(kind="bar", y="U.S.")
```

## Comparing with Crude Oil Production


```python
file = "/Users/elifcubukcu/data/U.S._crude_oil_production.csv.xls"
oil_df = pd.read_csv(file, decimal=",", parse_dates=["Month"])
oil_df.head(3)
oil_df.columns
oil_df
```

```python
month_type = oil_df['Month'].dtype
month_type
```

```python
value_cols = oil_df.columns.drop("Month")
oil_df[value_cols] = oil_df[value_cols].apply(pd.to_numeric, errors="coerce")
```

---

## Yearly Oil production

```python
yearly_oil_df = oil_df.groupby(oil_df["Month"].dt.year)[value_cols].sum()
yearly_oil_df
```

```python
yearly_oil_df.columns
```

```python
yearly_oil_df.columns = yearly_oil_df.columns.str.strip()
yearly_oil_df.columns
```

```python
yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind='bar')
```

```python
years = yearly_oil_df.index

filtered_yearly_oil_df = yearly_oil_df[
    np.logical_and(
        years != years.min(),
        years != years.max()
    )
]
filtered_yearly_oil_df
```

```python
filtered_yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind='bar')
```

### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('oil',
    filtered_oil_shape=filtered_yearly_oil_df.shape,
    filtered_oil_index_year=filtered_yearly_oil_df.index[0],
    us_total=filtered_yearly_oil_df.iloc[0,0]
)
result.write()

print(result.check())
```
